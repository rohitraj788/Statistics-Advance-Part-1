1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random process or experiment. It assigns numerical values to each outcome in a sample space.

2. What are the types of random variables?

- Discrete random variable: Takes countable values (e.g., 0, 1, 2, …).

- Continuous random variable: Takes values from an interval (uncountable, like real numbers).

3. What is the difference between discrete and continuous distributions?

- Discrete distributions describe random variables with distinct outcomes (e.g., binomial, Poisson).

- Continuous distributions describe variables with outcomes over a continuous range (e.g., normal, uniform).

4. What are probability distribution functions (PDF)?

For continuous variables, a Probability Density Function (PDF) gives the relative likelihood of a variable falling within an interval. The area under the curve equals 1.

5. How do cumulative distribution functions (CDF) differ from PDF?

- PDF: Likelihood density at a point (for continuous variables).

- CDF: Cumulative probability that a variable is less than or equal to a value, for both discrete and continuous variables.



6. What is a discrete uniform distribution?

A distribution where each outcome is equally likely. Example: Rolling a fair die (1–6) — each has a 1/6 probability.

7. What are the key properties of a Bernoulli distribution?

- Only two outcomes: success (1) or failure (0).

- Defined by one parameter: p, the probability of success.

- Mean = p, Variance = p(1−p).

8. What is the binomial distribution, and how is it used in probability?

It models the number of successes in n independent Bernoulli trials with the same probability p.
Used in quality control, surveys, etc.

9. What is the Poisson distribution and where is it applied?

It models the number of events in a fixed interval of time or space when events occur with a constant mean rate.
Used in modeling arrival rates, e.g., calls to a call center.

11. What are the characteristics of a normal distribution?

- Bell-shaped and symmetric

- Defined by mean (μ) and standard deviation (σ)

- 68-95-99.7 Rule: Approx. 68% of data lies within 1σ of μ, etc.

12. What is the standard normal distribution, and why is it important?

A normal distribution with mean 0 and standard deviation 1.
Used to standardize values and compute probabilities using Z-scores.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population’s distribution.

14. How does the Central Limit Theorem relate to the normal distribution?

It justifies the use of normal approximation in statistics — crucial for hypothesis testing and confidence intervals when the sample size is large.

15. What is the application of Z statistics in hypothesis testing?

Z-statistics are used to determine how many standard deviations a sample mean is from the population mean. Helps calculate p-values and decide whether to reject a null hypothesis.

17. What are point estimates and interval estimates in statistics?

Point estimate: A single value (e.g., sample mean) used to estimate a parameter.

Interval estimate: A range (e.g., confidence interval) around the point estimate that is likely to contain the true parameter.

18. What is the significance of confidence intervals in statistical analysis?

They provide a range where the population parameter is likely to lie with a certain confidence level (e.g., 95%). Offers more information than a point estimate.

19. What is the relationship between a Z-score and a confidence interval?

Z-scores determine the critical values used to build confidence intervals for population parameters when the standard deviation is known.

20. How are Z-scores used to compare different distributions?

They standardize scores from different distributions, allowing meaningful comparisons (e.g., comparing test scores from different exams).

21. What are the assumptions for applying the Central Limit Theorem?

- Independent and identically distributed (i.i.d) samples

- Sample size is large enough (n ≥ 30 typically)

- Finite mean and variance of the population

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0
    return p if x == 1 else (1 - p)

# Example usage
print("P(X=1):", bernoulli_pdf(1, 0.6))
print("P(X=0):", bernoulli_pdf(0, 0.6))
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

9. Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(x - mean) / std for x in data]

data = np.random.randint(10, 100, 100)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y)
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.xlabel("Z")
    plt.ylabel("PDF")
    plt.grid(True)
    plt.show()

plot_standard_normal()
```

18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
from scipy.stats import norm

def z_score(x, mean, std):
    return (x - mean) / std

x_val = 75
mean = 70
std = 5
z = z_score(x_val, mean, std)
p = norm.cdf(z)

print(f"Z-score: {z:.2f}")
print(f"Probability from standard normal (P(Z < {z:.2f})) = {p:.4f}")
```

19. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

sample = np.random.normal(loc=52, scale=5, size=50)
pop_mean = 50
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")
```

20. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import norm

data = np.random.normal(loc=100, scale=15, size=40)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
z = norm.ppf(0.975)  # 95% confidence

margin = z * (std / np.sqrt(n))
ci = (mean - margin, mean + margin)

print(f"95% Confidence Interval: {ci}")
```

21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
from scipy.stats import norm

data = np.random.normal(loc=60, scale=10, size=100)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

z = norm.ppf(0.975)
margin_error = z * (std / np.sqrt(n))
lower, upper = mean - margin_error, mean + margin_error

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({lower:.2f}, {upper:.2f})")
```

26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

```python
import numpy as np

data = np.random.normal(100, 20, 1000)
data_range = np.max(data) - np.min(data)
q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25

print(f"Range = {data_range:.2f}")
print(f"IQR = {iqr:.2f}")
```

27. Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(70, 10, 100)
z_scores = (data - np.mean(data)) / np.std(data)

plt.hist(data, bins=20, alpha=0.6, label='Original')
plt.hist(z_scores, bins=20, alpha=0.6, label='Z-normalized')
plt.legend()
plt.title("Z-score Normalization")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
